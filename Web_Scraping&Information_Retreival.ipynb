**Q.** Given a web document such as Email/Contact details, write a program to extract the following using Regular Expression:

a. Extract Phone Number(10 digit number and other variants)

b. Extract Email-id

c. Extract date

```python
import re   # for regular expressions
import requests  # for opening web page
import pandas as pd
```

```python
# Extracting phone numbers

page_url = 'https://www.jnu.ac.in/scss-contact-us'
print("Opening URL:", page_url)
page_data = requests.get(page_url)
page_html = str(page_data.content)

matches = []
matchgrp = ['[6-9]{1}[0-9]{9}', '\+\d{1,2}[-][6-9]{1}[0-9]{9}', '\+[\s]\d{1,2}[-][0-9]{2}[-][0-9]{8}', '\+\d{1,2}[-][0-9]{2}[-][0-9]{8}']

for j in matchgrp:
   pat = j
   result = re.findall(pat, page_html)
   if result:
      matches.append(result)

for i in matches:
  print(i)
```

```python
# Extracting email ids

page_url = 'https://www.jnu.ac.in/index.php/scss-faculty'
print("Opening URL:", page_url)
page_data = requests.get(page_url)
page_html = str(page_data.content)

matches = []
matchgrp = ['[A-Za-z0-9._]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}']

for j in matchgrp:
   pat = j
   result = re.findall(pat, page_html)
   if result:
      matches.append(result)

for i in matches:
  for j in i:
    print(j)
```

```python
# Extracting dates

page_url = 'https://www.wbpscupsc.com/timeline-of-indian-history-modern-india/'
print("Opening URL:", page_url)
page_data = requests.get(page_url)
page_html = str(page_data.content)

matches = []
matchgrp = ['\d{2}/\d{2}/\d{4}', '\d{4}-\d{2}-\d{2}', '\d{2}-\d{2}-\d{2}', '[A-Za-z]+\s\d{1,2},\s\d{4}']

for j in matchgrp:
   pat = j
   result = re.findall(pat, page_html)
   if result:
      matches.append(result)

for i in matches:
  print(i)
```

```python
import math
from collections import defaultdict
```

```python
# Calculate term frequency (TF) of a word in a document
def calculate_tf(word, document):
    word_count = document.count(word)
    total_words = len(document.split())
    return word_count / total_words

# Calculate inverse document frequency (IDF) of a word in the document set
def calculate_idf(word, documents):
    num_documents_with_word = sum(1 for doc in documents if word in doc)
    return math.log(len(documents) / (1 + num_documents_with_word))

# Calculate document frequency (DF) of a word in the document set
def calculate_df(word, documents):
    num_documents_with_word = sum(1 for doc in documents if word in doc)
    return num_documents_with_word
```

```python
# Calculate TF weights
def calculate_tf_weights(documents):
    tf_weights = defaultdict(dict)
    for doc_index, document in enumerate(documents):
        for word in document.split():
            tf = calculate_tf(word, document)
            tf_weights[doc_index][word] = tf
    return tf_weights


# Calculate TF*IDF weights
def calculate_tfidf_weights(documents):
    tfidf_weights = defaultdict(dict)
    for doc_index, document in enumerate(documents):
        for word in document.split():
            tf = calculate_tf(word, document)
            idf = calculate_idf(word, documents)
            tfidf_weights[doc_index][word] = tf * idf
    return tfidf_weights


# Calculate DF weights
def calculate_df_weights(documents):
    df_weights = defaultdict(dict)
    for doc_index, document in enumerate(documents):
        for word in document.split():
            df = calculate_df(word, documents)
            df_weights[doc_index][word] = df
    return df_weights


# Calculate TF*DF weights
def calculate_tfdf_weights(documents):
    tfdf_weights = defaultdict(dict)
    for doc_index, document in enumerate(documents):
        for word in document.split():
            tf = calculate_tf(word, document)
            df = calculate_df(word, documents)
            tfdf_weights[doc_index][word] = tf * df
    return tfdf_weights


# Store weights in dictionary
tf_weights = calculate_tf_weights(documents)
df_weights = calculate_df_weights(documents)
tfidf_weights = calculate_tfidf_weights(documents)
tfdf_weights = calculate_tfdf_weights(documents)
```

**Q.5.** Take a text corpus on certain topic. Do the preprocessing required and come up with a
sensible vocabulary. Now represent each document in the dimension of the vocabulary.
Find out Frequent patterns of size 1,2,3,4,5 using apriori algorithm.(to be implemented by

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Preprocessing function
def preprocess_text(text):
    # Tokenize
    tokens = word_tokenize(text.lower())

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [token for token in tokens if token not in stop_words]

    # Lemmatize the words
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Join the tokens back into a single string
    preprocessed_text = ' '.join(lemmatized_tokens)

    return preprocessed_text
```

```python
corpus = [pdf1, pdf2, pdf3]

preprocessed_corpus = [preprocess_text(doc) for doc in corpus]
```

```python
vectorizer = CountVectorizer()

vectorizer.fit(preprocessed_corpus)

vocabulary = vectorizer.get_feature_names_out()
```

```python
doc_term_matrix = vectorizer.transform(preprocessed_corpus)

# Convert the document-term matrix to an array
doc_term_array = doc_term_matrix.toarray()

# Print the document-term matrix for each document
for i, doc_vector in enumerate(doc_term_array):
    print(f"Document {i + 1}:")
    for j, term_count in enumerate(doc_vector):
        if term_count > 0:
            print(f"  {vocabulary[j]}: {term_count}")
```
